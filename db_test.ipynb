```python
def _query(sql: str = None):
    db, cursor = connect_to_db()
    if sql is None:
        sql = """
            SELECT *
            FROM quantus.quantus_license
        """

    cursor.execute(sql)
    result = cursor.fetchall()

    cursor.close()
    db.close()
    
    return result

def get_all(table):
    sql = f"""
        SELECT *
        FROM {table};
    """
    result = _query(sql)
    # result = _query()

    result_df = pd.DataFrame(result)

    return result_df
```

```python
def update(table: str, sets: dict, wheres: dict):
    set_fields = list(sets.keys())
    where_fields = list(wheres.keys())
    
    for set_field, where_field in zip(set_fields, where_fields):
        sql = f"""
            UPDATE {table}
            SET {set_field}={sets[set_field]}
            WHERE {where_field}={wheres[where_field]};
        """
        result = _query(sql)
    
    return sql, result
```
